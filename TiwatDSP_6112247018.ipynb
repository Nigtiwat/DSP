```python
#Lab1
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

sin_wave=thinkdsp.SinSignal(freq=7018, amp=18, offset=0)
sin_wave.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#lab2
#TriangleSignal(คือค่า F)
#Triangle *18 เสียงจะยาวขึ้น
signal = thinkdsp.TriangleSignal(18)
duration = signal.period*18
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
wave = signal.make_wave(duration,framerate=10000)
wave.apodize() 
wave.make_audio()
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python

signal = thinkdsp.SquareSignal(18)
duration = signal.period*18
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
segment.plot()
wave = signal.make_wave(duration,framerate=10000)
wave.apodize() 
wave.make_audio()

```
